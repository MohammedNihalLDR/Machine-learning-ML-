```python
def get_continent(country):
    # Remove leading and trailing whitespace, including the invisible character
    country = country.strip()
    # Europe
    if country in ["Denmark","Sweden","Norway","Finland","Switzerland","Netherlands","Luxembourg",
                   "Iceland","Germany","Ireland","United Kingdom","Austria","Belgium","France",
                   "Spain","Czech Republic","Portugal","Slovenia","Malta","Italy","Latvia","Lithuania",
                   "Cyprus","Slovakia","Poland","Greece","Croatia","Hungary","Romania","Bulgaria",
                   "Montenegro","Serbia","Macedonia","Albania","Moldova","Bosnia and Herzegovina",
                   "Ukraine","Belarus","Estonia","Turkey", "Georgia", "Russia"]:
        return "Europe"

    # Asia
    elif country in ["Japan","Singapore","Taiwan","Hong Kong","South Korea","Israel","United Arab Emirates",
                     "Qatar","Malaysia","China","Kuwait","Armenia","Bahrain","Indonesia","Thailand",
                     "Kazakhstan","Vietnam","Saudi Arabia","Jordan","Mongolia","Sri Lanka","Azerbaijan",
                     "Kyrgyzstan","Uzbekistan","India","Turkmenistan","Nepal","Lebanon","Tajikistan",
                     "Laos","Afghanistan", "Oman", "Yemen", "Bangladesh", "Iran", "Pakistan", "Iraq",
                     "Myanmar", "Syria","Philippines"]:
        return "Asia"

    # Africa
    elif country in ["South Africa","Mauritius","Seychelles","Morocco","Ghana","Tunisia","Senegal",
                     "Kenya","Algeria","Rwanda","Gambia","Tanzania","Cambodia","Gabon","Côte d'Ivoire",
                     "Egypt","Djibouti","Benin","Malawi","Zambia","Comoros","Uganda","Equatorial Guinea",
                     "Lesotho","Togo","Swaziland","Burkina Faso","Madagascar","Liberia","Zimbabwe",
                     "Guinea","Nigeria","Sierra Leone","Libya","Cameroon","Congo","Guinea-Bissau",
                     "Mozambique","Mali","Niger","Ethiopia","Mauritania","Angola","Burundi","Sudan",
                     "Eritrea","Democratic Republic of Congo","Chad","Somalia","Central African Republic",
                     "Cabo Verde", "Botswana", "São Tomé and Príncipe", "Namibia", "South Sudan"]:
        return "Africa"

    # North America
    elif country in ["Canada","United States","Costa Rica","Panama","Trinidad and Tobago","Jamaica",
                     "Dominican Republic","Mexico","Belize","El Salvador","Guatemala","Cuba","Honduras",
                     "Haiti","Nicaragua"]:
        return "North America"

    # South America
    elif country in ["Chile","Uruguay","Argentina","Peru","Brazil","Colombia","Suriname","Ecuador",
                     "Paraguay","Guyana","Bolivia","Venezuela"]:
        return "South America"

    # Oceania
    elif country in ["New Zealand","Australia","Papua New Guinea"]:
        return "Oceania"

    else:
        return "Unknown"

# Apply the function
df["Continent"] = df["Country"].apply(get_continent)
```

# **Problem Statement**
*   In today’s globalized world, people have different goals when considering countries to visit, study, settle, or do business. Choosing the right country can be overwhelming because each country differs in safety, freedom, government quality, economy, health, education, infrastructure, and other factors. Often, users do not have the tools to compare countries effectively based on their personal priorities, leading to suboptimal choices. Therefore, there is a need for a system that can recommend countries based on a user’s specific requirements and goals, considering multiple factors simultaneously.




# **Justification**



*   This model is important because it provides personalized recommendations tailored to individual needs. Instead of showing all countries equally, it considers the user’s goal (Travel, Study, Settle, Business) and continent preference, focusing only on the relevant features for that goal. By using Nearest Neighbors with cosine similarity, the system identifies countries whose characteristics are most similar to the user’s inputs. The standardization of numeric features ensures fair comparison across features with different scales. Additionally, showing a match percentage helps the user understand how closely each country aligns with their preferences, making the decision process more transparent and logical.
# **The model is a goal-based country recommendation system. It allows the user to:**



*   Choose a goal (Travel, Study, Settle, Business)
*   Provide values for relevant features (like Safety, Economy, Education, Health, etc.) and select a continent







*   Receive the top 5 countries that best match their requirements, along with a percentage match

```python
# Goal-based relevant features
goal_features = {
    "Travel": ["Safety", "Freedom", "Government", "Society", "Infrastructure", "Living", "Health", "Nature", "Continent"],
    "Study": ["Safety", "Freedom", "Government", "Society", "Infrastructure", "Economy", "Living", "Health", "Education", "Continent"],
    "Settle": ["Safety", "Freedom", "Government", "Society", "Investment", "Business", "Infrastructure", "Economy", "Living", "Health", "Education", "Nature", "Continent"],
    "Business": ["Safety", "Freedom", "Government", "Investment", "Business", "Infrastructure", "Economy", "Living", "Health", "Continent"]
}
```

```python
# Mapping for continent options
continent_map = {0: "Africa", 1: "Asia", 2: "Europe", 3: "South America", 4: "North America", 5: "Oceania"}
```

```python
9# Ask user for goal
print("Choose your goal: Travel / Study / Settle / Business")
goal = input("Enter goal: ").capitalize()
if goal not in goal_features:
    print("Invalid goal! Defaulting to Travel.")
    goal = "Travel"

# Prepare numeric features for the goal (exclude Continent)
num_features_for_goal = [f for f in goal_features[goal] if f != "Continent"]

# Standardize numeric features for the entire dataset
scaler = StandardScaler()
df[num_features_for_goal] = scaler.fit_transform(df[num_features_for_goal])

# Get user input
user_input = {}
for feature in goal_features[goal]:
    if feature == "Continent":
        print("Select Continent:")
        for key, val in continent_map.items():
            print(f"{key} - {val}")
        while True:
            try:
                continent_index = int(input("Enter Continent number: "))
                if continent_index in continent_map:
                    continent_name = continent_map[continent_index]
                    user_input[feature] = le.transform([continent_name])[0]
                    break
                else:
                    print("Invalid number. Try again.")
            except:
                print("Invalid input. Enter a number 0-5.")
    else:
        while True:
            try:
                value = float(input(f"Enter value for {feature} (0-100): "))
                if 0 <= value <= 100:
                    user_input[feature] = value
                    break
                else:
                    print("Value must be between 0 and 100.")
            except:
                print("Invalid input. Enter a number.")

# Filter dataset by selected continent
df_filtered = df[df["Continent"] == user_input["Continent"]]

# Train Nearest Neighbors on numeric features of filtered dataset
model = NearestNeighbors(n_neighbors=3, metric='cosine')
model.fit(df_filtered[num_features_for_goal])

# Convert user input to DataFrame and standardize numeric features
user_df = pd.DataFrame([user_input])
user_df[num_features_for_goal] = scaler.transform(user_df[num_features_for_goal])

# Find top 5 nearest countries in that continent
distances, indices = model.kneighbors(user_df[num_features_for_goal])
recommended_countries = df_filtered.iloc[indices[0]]["Country"].values
similarity_percent = (1 - distances[0]) * 100  # convert cosine distance to percentage match

# Print recommendations with match percentage
print("\nRecommended Countries with Match Percentage:")
for country, percent in zip(recommended_countries, similarity_percent):
    print(f"{country}: {percent:.2f}% match")
```
